```python
#combine graph till now
x = np.arange(0,3*np.pi,0.5)
y_s = np.sin(x)
y_c = np.cos(x)
plt.subplot(2,1,1)
plt.title('sin graph',size = 20,color = 'blue')
plt.xlabel('x-axis',size= 20)
plt.ylabel('y-axis',size = 20)
plt.plot(x,y_s,'o-',color = 'green',label = 'sin')
plt.legend()
plt.subplot(2,1,2)
plt.title('cos graph',size = 20,color = 'purple')
plt.xlabel('x-axis',size = 20)
plt.ylabel('y-axis',size = 20)
plt.plot(x,y_c,'x-',color = 'red',label = 'cos')
plt.legend()
plt.show()
```

```python
#Bar Graph
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lang = ['c++','python','java','c','ruby']
students = [10,20,30,40,50]
ax.bar(lang,students)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
lang = ['c++','python','java','c','ruby']
students = [10,50,30,80,50]
ax.bar(lang,students,color = 'gold')
plt.title('Bar graph',size = 25,color = 'red')
plt.xlabel('language known',size = 20)
plt.ylabel('no of students',size = 20)
plt.show()
```

```python
lang = ['c++','python','java','c']
students = [10,50,30,80]
colors = ['green','gold','yellowgreen','red']
patches,texts = plt.pie(students,colors=colors)
plt.show()
```

```python
lang = ['c++','python','java','c']
students = [10,50,30,80]
colors = ['green','gold','yellowgreen','red']
patches,texts = plt.pie(students, colors=colors)
plt.legend(patches,lang)
plt.show()
```

```python
lang = ['c++','python','java','c']
students = [10,50,30,80]
colors = ['green','gold','yellowgreen','red']
patches,texts= plt.pie(students, colors=colors,startangle=90,shadow=True)
plt.legend(patches,lang)
plt.title('Pie graph',size =25,color='red')
plt.axis('equal')
plt.show()
```
