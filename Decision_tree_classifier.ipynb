# Decision Tree Classifier
------------------------------------------

- Decision tree is a supervised classification algorithm used for the purpose of classification and regression both.


- There is a algorithm **Cart** which is used for this classification purpose.

**using the iris dataset, we can construct the tree as follows:**

```python
from sklearn.datasets import load_iris
from sklearn import tree
X, y = load_iris(return_X_y = True)
```

```python
print(X)
```

```python
print(y)
```

**Decision tree classifier**

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X,y)
```

**once trained, we can plot the tree with the plot_tree function:**

```python
tree.plot_tree(clf)
```

```python
# !conda install python-graphviz
# !pip install graphviz
```

```python
import graphviz
dot_data = tree.export_graphviz(clf, out_file=None)
graph = graphviz.Source(dot_data)
graph.render("iris")
```

```python
iris = load_iris()
dot_data = tree.export_graphviz(clf, out_file=None, feature_names = iris.feature_names, class_names = iris.target_names, filled = True, rounded = True, special_characters = True)
graph = graphviz.Source(dot_data)
graph
```
