**groupby and aggregate**

```python
col = ['population','ocean_proximity']
data = df[col]

data = data.groupby('ocean_proximity').sum()
print(data)
# here sum is the aggregation function
```

```python
col = ['population','ocean_proximity']
data = df[col]
data = data.groupby('ocean_proximity').mean()
print(data)
```
